# ***Exploratory Data Analysis***

```python
# Set the figure size
plt.figure(figsize=(15, 6))
# Define a color palette for the bars
colors = ["#003366", "#99c2ff"]
# Create a bar plot
sns.barplot(data=df, x='Country', y='Age', hue='Gender', errorbar=None, palette=colors)
# Set the title
plt.title("Age Distribution By Country & Gender")
# Label the x-axis
plt.xlabel("Country")
# Label the y-axis
plt.ylabel("Age")
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Count the number of patients in each continent
continent_counts = df['Continent'].value_counts()
# Set the size of the figure
plt.figure(figsize=(12, 8))
# Define a color palette for the pie chart segments
colors = ["#99c2ff", "#66a3ff", "#3385ff", "#0066ff", "#0052cc", "#004099", "#cce0ff", "#e6f2ff"]
# Create a pie chart
plt.pie(continent_counts, labels=continent_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
# Set the title
plt.title('Patient Distribution By Continent')
# Display the plot
plt.show()
```

```python
# Create a pivot table to summarize blood pressure counts by continent and country
pivot_table = df.pivot_table(index='Continent', columns='Country', values='Blood Pressure', aggfunc='count')
# Set the figure size
plt.figure(figsize=(15, 6))
# Create a heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='g')
# Set the title
plt.title("Blood Pressure Count By Countries & Continent")
# Label the x-axis
plt.xlabel("Country")
# Label the y-axis
plt.ylabel("Continent")
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))
# Create a histogram
sns.histplot(data=df, x="Cholesterol", bins=20, kde=True, color="#66a3ff")
# Set the title
plt.title("Cholesterol Distribution")
# Label the x-axis
plt.xlabel("Cholesterol (mg/dL)")
# Label the y-axis
plt.ylabel("Frequency")
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(18, 6))
# Create a line plot
sns.lineplot(data=df, x="Country", y="Exercise Hours Per Week", marker="o", color="#003366")
# Set the title
plt.title("Exercise Hours Per Week By Country")
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))
# Define a color palette for the donut chart
color = ["#66a3ff", "#0066ff", "#004099"]
# Count the occurrences of each diet type in the dataset
diet_counts = df["Diet"].value_counts()
# Create a donut chart
plt.pie(diet_counts, labels=diet_counts.index, autopct="%1.1f%%", colors=color, wedgeprops={"width": 0.4})
# Set the title
plt.title("Diet Distribution")
# Display the chart
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(18, 6))
# Create a line plot
sns.lineplot(data=df, x="Age", y="Heart Rate", marker="o", color="#00224d")
# Set the title
plt.title("Heart Rate By Age")
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))
# Create a count plot
sns.countplot(data=df, x='Alcohol Consumption', hue='Continent', palette="Blues")
# Set the title
plt.title('Alcohol Consumption By Continent')
# Label the x-axis
plt.xlabel("Alocohol Consumption")
# Label the y-axis
plt.ylabel("Count")
# Display the plot
plt.show()
```

```python
# Create a pivot table to summarize BMI  by continent and country
pivot_table = df.pivot_table(index='Continent', columns='Country', values='BMI', aggfunc='mean')
# Set the figure size
plt.figure(figsize=(15, 6))
# Create a heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f')
# Set the title
plt.title("Average BMI By Continent & Country")
# Label the x-axis
plt.xlabel("Country")
# Label the y-axis
plt.ylabel("Continent")
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))
# Create a histogram
sns.histplot(data=df, x='Age', hue='Smoking', multiple='stack', bins=10, palette='Blues')
# Set the title
plt.title('Age Distribution By Smoking Status')
# Display the plot
plt.show()
```

```python
# Count the occurrences of each smoking status in the dataset
smoking_counts = df['Obesity'].value_counts()
# Define a color palette for the donut chart
color = ["#0066ff", "#004099"]
# Set the figure size
plt.figure(figsize=(8, 6))
# Create a donut chart
plt.pie(smoking_counts, labels=smoking_counts.index, colors=color, autopct='%1.1f%%', startangle=90,wedgeprops=dict(width=0.4))
# Set the title
plt.title('Obesity Status Distribution')
# Display the chart
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(18, 6))
# Group the data by year and calculate the average exercise hours
average_exercise = df.groupby('Country')['Heart Attack Risk'].mean().reset_index()
# Create an area plot to visualize the trend of average exercise hours over the years
sns.lineplot(data=average_exercise, x='Country', y='Heart Attack Risk', marker='o', color='blue')
plt.fill_between(average_exercise['Country'], average_exercise['Heart Attack Risk'], color='blue', alpha=0.3)
# Set the title
plt.title('Average Heart Attack Risk By Countries')
# Label the x-axis
plt.xlabel('Country')
# Label the y-axis
plt.ylabel('Average Heart Attack Risk')
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Group the data by Country and calculate the mean stress level for each country
country_stress = df.groupby('Country')['Stress Level'].mean().reset_index()
# Set the figure size
plt.figure(figsize=(15, 6))
# Create a line plot
sns.lineplot(data=country_stress, x="Country", y="Stress Level", marker="o", color="#00224d")
# Set the title
plt.title("Stress Level By Country")
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```
