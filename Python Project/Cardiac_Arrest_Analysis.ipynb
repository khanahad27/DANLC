# ***Cardiac Arrest Analysis***

```python
# Set the figure size for the plot
plt.figure(figsize=(15, 6))
# Define a color palette for the bars
colors = ["#003366", "#99c2ff"]
# Create a bar plot showing the average age distribution by country and gender
sns.barplot(data=df, x='Country', y='Age', hue='Gender', errorbar=None, palette=colors)
# Set the title of the plot
plt.title("Age Distribution by Country and Gender")
# Label the x-axis
plt.xlabel("Country")
# Label the y-axis
plt.ylabel("Age")
# Rotate the x-axis labels for improved readability
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Count the number of patients in each continent
continent_counts = df['Continent'].value_counts()
# Set the size of the figure for the pie chart
plt.figure(figsize=(10, 6))
# Define a color palette for the pie chart segments
colors = ["#99c2ff", "#66a3ff", "#3385ff", "#0066ff", "#0052cc", "#004099", "#cce0ff", "#e6f2ff"]
# Create a pie chart to visualize the distribution of patients by continent
plt.pie(continent_counts, labels=continent_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
# Set the title for the pie chart
plt.title('Patient Distribution by Continent')
# Ensure the pie chart is circular by setting equal aspect ratio
plt.axis('equal')
# Display the pie chart
plt.show()
```

```python
# Create a pivot table to summarize blood pressure counts by continent and country
pivot_table = df.pivot_table(index='Continent', columns='Country', values='Blood Pressure', aggfunc='count')
# Set the figure size for the heatmap
plt.figure(figsize=(15, 6))
# Create a heatmap to visualize the blood pressure data across countries and continents
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='g')
# Set the title for the heatmap
plt.title("Heatmap of Blood Pressure (Countries by Continent)")
# Label the x-axis
plt.xlabel("Country")
# Label the y-axis
plt.ylabel("Continent")
# Display the heatmap
plt.show()
```

```python
# Set the figure size for the histogram to ensure clarity in visualization
plt.figure(figsize=(12, 6))
# Define a color palette for the histogram
colors = ["#00224d", "#003366", "#99c2ff", "#66a3ff", "#3385ff", "#0066ff", "#0052cc", "#004099", "#cce0ff", "#e6f2ff"]
# Apply the defined color palette to the seaborn plots
sns.set_palette(colors)
# Create a histogram to visualize the distribution of cholesterol levels across different age groups
plt.figure(figsize=(10, 6))
# Use seaborn's histplot to plot the age distribution with a kernel density estimate (KDE) overlay
sns.histplot(data=df, x="Age", bins=20, kde=True)
# Set the title for the histogram
plt.title("Distribution of Cholesterol Levels")
# Label the x-axis to indicate the variable being measured
plt.xlabel("Age")
# Label the y-axis to indicate the measurement unit for cholesterol
plt.ylabel("Cholesterol (mg/dL)")
# Display the histogram
plt.show()
```

```python
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(12, 6))
# Define a color for the line plot to maintain consistency in the visual theme
colors = ["#00224d", "#003366"]
# Create a line plot to show the average exercise hours per week for each country
sns.lineplot(data=df, x="Country", y="Exercise Hours Per Week", marker="o", color=colors[0])
# Set the title for the line plot to describe the data being visualized
plt.title("Exercise Hours Per Week by Country")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```

```python
# Set the figure size for the donut chart to ensure it is visually appealing
plt.figure(figsize=(12, 6))
# Define a color palette for the donut chart
colors = ["#99c2ff", "#66a3ff", "#3385ff", "#0066ff", "#0052cc", "#004099", "#cce0ff", "#e6f2ff"]
# Apply the defined color palette to the seaborn plots
sns.set_palette(colors)
# Count the occurrences of each diet type in the dataset
diet_counts = df["Diet"].value_counts()
# Create a donut chart to visualize the distribution of diet types among the population
plt.pie(diet_counts, labels=diet_counts.index, autopct="%1.1f%%", colors=colors[:len(diet_counts)],
        wedgeprops={"width": 0.4})
# Set the title for the donut chart to describe the data being visualized
plt.title("Diet Distribution")
# Display the donut chart
plt.show()
```

```python
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(18, 6))
# Create a line plot to show the relationship between age and heart rate
sns.lineplot(data=df, x="Age", y="Heart Rate", marker="o", color="#00224d")
# Set the title for the line plot to describe the data being visualized
plt.title("Heart Rate by Age")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```

```python
# Set the figure size for the bar plot to ensure clarity in visualization
plt.figure(figsize=(10, 6))
# Create a bar plot to visualize alcohol consumption categorized by continent
sns.countplot(data=df, x='Alcohol Consumption', hue='Continent')
# Set the title for the bar plot to describe the data being visualized
plt.title('Alcohol Consumption by Continent')
# Display the bar plot
plt.show()
```

```python
# Create a pivot table to summarize BMI data by continent and country
pivot_table = df.pivot_table(index='Continent', columns='Country', values='BMI', aggfunc='mean')
# Set the figure size for the heatmap to enhance visualization
plt.figure(figsize=(15, 6))
# Create a heatmap to visualize the average BMI data across countries and continents
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f')
# Set the title for the heatmap to describe the data being visualized
plt.title("Heatmap of Average BMI by Country and Continent")
# Label the x-axis to indicate the countries
plt.xlabel("Country")
# Label the y-axis to indicate the continents
plt.ylabel("Continent")
# Display the heatmap
plt.show()
```

```python
# Set the figure size for the histogram to enhance visibility
plt.figure(figsize=(12, 6))
# Create a histogram to visualize the age distribution categorized by smoking status
sns.histplot(data=df, x='Age', hue='Smoking', multiple='stack', bins=10)
# Set the title for the histogram to describe the data being visualized
plt.title('Age Distribution by Smoking Status')
# Display the histogram
plt.show()
```

```python
# Count the occurrences of each smoking status in the dataset
smoking_counts = df['Obesity'].value_counts()
# Set the figure size for the donut chart to ensure it is visually appealing
plt.figure(figsize=(8, 4))
# Create a donut chart to visualize the distribution of smoking status
plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%', startangle=90,
        wedgeprops=dict(width=0.3))
# Set the title for the donut chart to describe the data being visualized
plt.title('Obesity Status Distribution')
# Display the donut chart
plt.show()
```

```python
# Set the figure size for the area plot to enhance visibility
plt.figure(figsize=(18, 6))
# Group the data by year and calculate the average exercise hours
average_exercise = df.groupby('Country')['Heart Attack Risk'].mean().reset_index()
# Create an area plot to visualize the trend of average exercise hours over the years
sns.lineplot(data=average_exercise, x='Country', y='Heart Attack Risk', marker='o', color='blue')
plt.fill_between(average_exercise['Country'], average_exercise['Heart Attack Risk'], color='blue', alpha=0.3)
# Set the title for the area plot to describe the data being visualized
plt.title('Average Heart Attack Risk Over Countries')
# Label the x-axis to indicate the years
plt.xlabel('Country')
# Label the y-axis to indicate the average exercise hours
plt.ylabel('Average Heart Attack Risk')
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the area plot
plt.show()
```

```python
# Group the data by 'Country' and calculate the mean stress level for each country
country_stress = df.groupby('Country')['Stress Level'].mean().reset_index()
# Set the figure size for the line plot to enhance visibility
plt.figure(figsize=(15, 6))
# Create a line plot to visualize the average stress level for each country
sns.lineplot(data=country_stress, x="Country", y="Stress Level", marker="o", color="#00224d")
# Set the title for the line plot to describe the data being visualized
plt.title("Stress Level by Country")
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
# Display the line plot
plt.show()
```
