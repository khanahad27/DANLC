```python
# Import necessary libraries
import mysql.connector
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Connect to MySQL
db = mysql.connector.connect(
    host="localhost",       # Replace with your host (e.g., "127.0.0.1")
    user="root",   # Your MySQL username
    password="271045", # Your MySQL password
    database="project" # Optional: specify database to connect to
)

# Check connection
if db.is_connected():
    print("Connected to MySQL!")
```

```python
# Create cursor object to execute sql queries
cursor = db.cursor()
# Query the data from cybersecurity attacks
cursor.execute("SELECT * FROM data")
# After fetching data from the database we are storing it into Pandas DataFrame
df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
# Print the Data
print(df)
```

# Exploratory Data Analysis

```python
# Count the occurrences of each grade level in the DataFrame
traffic_counts = df['Grade_Level'].value_counts()
# Set the figure size 
plt.figure(figsize=(10, 6))
# Define a color palette for the pie chart segments
colors = ["#8B0000","#FDD4D8", "#F5A68D", "#B84349"]
# Create a pie chart 
plt.pie(traffic_counts, labels=traffic_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
# Set the title 
plt.title('Grade Level Distribution')
# Ensure the pie chart is a circle by setting the aspect ratio to equal
plt.axis('equal')
# Display the plot
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a count plot with ordered months
sns.countplot(data=df, x='Age', hue='Gender', palette='Reds')
# Set the title 
plt.title("Age Distribution By Gender")
# Label the x-axis 
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Count")
# Display the plot
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a histogram 
sns.histplot(data=df, x="English_Score", bins=20, kde=True, color="#FDD4D8")
# Set the title 
plt.title("English Score Distribution")
# Label the x-axis
plt.xlabel("English Score")
# Label the y-axis
plt.ylabel("Frequency")
# Display the histogram
plt.show()
```

```python
# Create a pivot table to summarize the count of math score by grade level and age
pivot_table = df.pivot_table(index="Grade_Level", columns="Age", values="Math_Score", aggfunc="mean")
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a heatmap 
sns.heatmap(pivot_table, annot=True, fmt='g', cmap="coolwarm")
# Set the title
plt.title("Average Math Score By Grade Level & Age")
# Label the x-axis
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Grade Level")
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(18, 6))
# Create a line plot 
sns.lineplot(data=df, x="Age", y="Study_Hours_Per_Week", marker="o", color="#B84349")
# Set the title 
plt.title("Study Hours Per Week By Age")
# Rotate the x-axis labels 
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Define a color palette for the donut chart
color = ["#FDD4D8", "#F5A68D", "#B84349"]
# Count the occurrences of each action taken in the dataset
action_counts = df["Parental_Involvement"].value_counts()
# Create a donut chart 
plt.pie(action_counts, labels=action_counts.index, autopct="%1.1f%%", colors=color, wedgeprops={"width": 0.4})
# Set the title
plt.title("Parental Involvement Distribution")
# Display the chart
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a bar plot 
sns.countplot(data=df, x="Extracurricular_Activities", hue="Gender", palette="Reds")
# Set the title for the bar plot to describe the data being visualized
plt.title("Extracurricular Activities By Gender")
# Label the x-axis
plt.xlabel("Extracurricular Activities")
# Label the y-axis
plt.ylabel("Count")
# Display the plot
plt.show()
```

```python
# Create a pivot table to summarize user information data by year and month
pivot_table = df.pivot_table(index="Gender", columns="School_Funding_Level", values="Age", aggfunc="count")
# Set the figure size
plt.figure(figsize=(20, 6))
# Create a heatmap
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="coolwarm")
# Set the title
plt.title("Count Of Age By Gender and School Funding Level")
# Label the x-axis
plt.xlabel("School Funding Level")
# Label the y-axis
plt.ylabel("Gender")
# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(18, 6))
# Create a line plot 
sns.lineplot(data=df, x="Age", y="Participation_Rate", marker="o", color="#00224d")
# Set the title
plt.title("Participation Rate By Age")
# Rotate the x-axis labels 
plt.xticks(rotation=45)
# Label the x-axis
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Participation Rate")
# Display the plot
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a histogram 
sns.histplot(data=df, x="Household_Income", hue="Scholarship_Status",bins=20, kde=True, color="#FDD4D8")
# Set the title 
plt.title("Household Income Distribution By Scholarship Status")
# Label the x-axis
plt.xlabel("Household Income")
# Label the y-axis
plt.ylabel("Frequency")
# Display the histogram
plt.show()
```

```python
# Set the figure size 
plt.figure(figsize=(12, 6))
# Create a bar plot 
sns.countplot(data=df, x="Age", hue="Technology_Access", palette="Blues")
# Set the title 
plt.title("Technology Access Distribution By Age")
# Label the x-axis
plt.xlabel("Age")
# Label the y-axis
plt.ylabel("Frequency")
# Display the plot
plt.show()
```

```python
# Count the occurrences of each attack type in the dataset
attack_type_counts = df["Future_Plans"].value_counts()
# Define a color palette for the donut chart
color = ["#8B0000","#FDD4D8", "#F5A68D", "#B84349"]
# Set the figure size
plt.figure(figsize=(8, 6))
# Create a donut chart
plt.pie(attack_type_counts, labels=attack_type_counts.index, autopct="%1.1f%%", colors=color, wedgeprops={"width": 0.4})
# Set the title 
plt.title("Future Plans Distribution")
# Display the chart
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 6))
# Group by 'States' and count unique 'Source IP Address'
states_source_ips = df.groupby("Grade_Level")["Attendance_Rate"].count().reset_index(name='Rate')
# Create an area plot to visualize the trend of unique IP source over the states
sns.lineplot(data=states_source_ips, x="Grade_Level", y="Rate", marker="o", color="blue")
plt.fill_between(states_source_ips["Grade_Level"], states_source_ips["Rate"], color="blue", alpha=0.3)
# Set the title 
plt.title("Attendance Rate By Grade Level")
# Label the x-axis
plt.xlabel("Grade Level")
# Label the y-axis
plt.ylabel("Attendance Rate")
# Rotate the x-axis labels
plt.xticks(rotation=45)
# Display the plot
plt.show()
```
